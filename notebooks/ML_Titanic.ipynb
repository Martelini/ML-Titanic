#Reading train and test data

```python
import pandas as pd
import numpy as np

np.random.seed(123)

final_test = pd.read_csv('https://raw.githubusercontent.com/Martelini/ML-Titanic/refs/heads/main/Data/test.csv')
data = pd.read_csv('https://raw.githubusercontent.com/Martelini/ML-Titanic/refs/heads/main/Data/train.csv')
data.head(15)
```

#Data cleaning

```python
#Drop nan in age

print('NaN for each column:')
print(data.isna().sum())

data = data.dropna(subset='Age')
data.head(10)
```

```python
x = data[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
y = data['Survived']
x.head()
```

```python
x = pd.get_dummies(x, columns=['Sex','Embarked'], dtype=int)
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

raw_train_x, raw_test_x, train_y, test_y = train_test_split(x, y, test_size=0.25)

scaler = StandardScaler()
scaler.fit(raw_train_x)
train_x = scaler.transform(raw_train_x)
test_x = scaler.transform(raw_test_x)
```

#Training models
##Linear SVC

```python
from sklearn.svm import LinearSVC

l_svc_model = LinearSVC()
l_svc_model.fit(train_x, train_y)

l_svc_result = l_svc_model.predict(test_x)
```

##SVC

```python
from sklearn.svm import SVC

svc_model = SVC(gamma='auto')
svc_model.fit(train_x, train_y)

svc_result = svc_model.predict(test_x)
```

##KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier()
knn_model.fit(train_x, train_y)

knn_result = knn_model.predict(test_x)
```

#Metrics

```python
from sklearn.metrics import accuracy_score

print('Accuracies: \n')

l_svc_acc = accuracy_score(l_svc_result, test_y)
print(f'Linear SVC: {l_svc_acc*100:.2f}%')

svc_acc = accuracy_score(svc_result, test_y)
print(f'SVC: {svc_acc*100:.2f}%')

knn_acc = accuracy_score(knn_result, test_y)
print(f'KNN: {knn_acc*100:.2f}%')
```
